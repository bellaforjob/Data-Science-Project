```python
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np

d1 = pd.read_csv('2019-Oct-simple.csv')
d2 = pd.read_csv('2019-Nov-simple.csv')
data_out = 'output.csv'

with open(data_out, 'w') as output:
    with open('2019-Oct-simple.csv', 'r')as input:
            line = input.readline()
            while line:
                output.write(line)
                line = input.readline()

    with open('2019-Nov-simple.csv','r')as input:
            line = input.readline()
            while line:
                output.write(line)
                line = input.readline()
```

```python
#load the file, read it
data = pd.read_csv(data_out)
```

```python
# sort the data on 3 columns
df = data.sort_values(['event_type','category_code', 'brand'])

# Find missing data in category_code and brand columns,  and fill it with nan
df.category_code[df['category_code']=='']=np.nan
df.brand[df['brand'] == ''] = np.nan
df.head(-100)
```

```python
#remove the row if nan in columns of category_code or brand
df = df.dropna()
        
df.head(-500)
```
