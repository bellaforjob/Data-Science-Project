```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as pp
```

```python
#load the file, read it
d1 = pd.read_csv('2019-Oct-m.csv')
d2 = pd.read_csv('2019-Nov-m.csv')
```

```python
# Data wrangling for d1
d1.isnull().any()
```

```python
#check the information of d1
d1.info()
```

```python
# sort the data on 2 columns that contains null values
d1.sort_values(['category_code', 'brand'])
d1
```

```python
#remove the row which missing category_code and brand
d1.dropna(how='all', axis=0, inplace = True)
d1.tail()
```

```python
# sort the data by product_id and only take columns we need
dn = d1[['event_time','event_type','product_id','category_code','brand','price','user_id']].sort_values(['product_id'])
dn.head()
```

```python
#fill missing values by the same columns's value based on the same product_id 
for index,row in dn.iterrows():
    if row['category_code'] == np.nan and row['brand']==np.nan:
        pid = row.product_id
        cate = row.category_code
        brand = row.brand
    elif row['category_code']==np.nan and row['brand']!=np.nan:
        row['category_code'] = row.groupby([pid, brand])[cate].bfill()
    elif row['brand']==np.nan and row['category_code']!=np.nan:
        row['brand'] = row.groupby([pid, cate])[brand].bfill()

dn.tail()
```

```python
#if there is not the same data to be copy in null value, fill null value as 'non-avaliable'
dn = dn.fillna('non-avaliable')
        
dn.tail(100)
```
